```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow import keras
from keras.utils import to_categorical
import matplotlib.pyplot as plt
from sklearn import metrics
```

```python
# (X_train, Y_train), (X_test, Y_test) = mnist.load_data()
# data_X = np.concatenate([X_train, X_test])
# data_y = np.concatenate([Y_train, Y_test])
(X_train, y_train), (X_test, y_test) = mnist.load_data()
print(y_train.shape,y_test.shape)
```

```python
temp = []
for i in range(len(y_train)):
    temp.append(to_categorical(y_train[i], num_classes=10))
y_train = np.array(temp)

temp = []
for i in range(len(y_test)):
    temp.append(to_categorical(y_test[i], num_classes=10))
y_test = np.array(temp)
```

```python
activation_funcs = ['relu','tanh','sigmoid']

# 2 hidden layers
max_acc =0
print("2 layer networks")
for i in activation_funcs:
    for j in activation_funcs:
        model2 = keras.Sequential([
            keras.layers.Reshape(target_shape=(28 * 28,), input_shape=(28, 28)),
            keras.layers.Dense(units=100, activation=i),
            keras.layers.Dense(units=50, activation=j),
            keras.layers.Dense(units=10, activation='softmax')
        ])

        model2.compile(loss='categorical_crossentropy', 
                      optimizer='adam',
                      metrics=['acc'])
        
        print("i is ",i)
        print("j is",j)
        
        model2.fit(X_train, y_train, epochs=5, 
          validation_data=(X_test,y_test))
        
        y_pred = model2.predict(X_test)

        y_pred = np.argmax(y_pred,axis=1)
        y_test_conf = y_test
        y_test_conf = np.argmax(y_test_conf,axis=1)

        con_mat = tf.math.confusion_matrix(labels=y_test_conf, predictions=y_pred).numpy()
        #print(con_mat)
        acc = model2.evaluate(X_test, y_test)[1]
        cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = con_mat)
        cm_display.plot(cmap='Oranges')
        plt.show()
        print('accuracy : ',acc)
        max_acc = max(acc,max_acc)
        print("\n\n\n\n\n\n")


        

print("max acc with 2 layer networks is ", max_acc)   

print("\n\n\n\n\n\n")
# for 3 hidden layers


print("3 layer networks")
max_acc =0
for i in activation_funcs:
    for j in activation_funcs:
        model2 = keras.Sequential([
            keras.layers.Reshape(target_shape=(28 * 28,), input_shape=(28, 28)),
            keras.layers.Dense(units=50, activation=i),
            keras.layers.Dense(units=50, activation=i),
            keras.layers.Dense(units=50, activation=j),
            keras.layers.Dense(units=10, activation='softmax')
        ])

        model2.compile(loss='categorical_crossentropy', 
                      optimizer='adam',
                      metrics=['acc'])
        
        print("i is ",i)
        print("j is",j)
        
        model2.fit(X_train, y_train, epochs=5, 
          validation_data=(X_test,y_test))
        
        y_pred = model2.predict(X_test)

        y_pred = np.argmax(y_pred,axis=1)
        y_test_conf = y_test
        y_test_conf = np.argmax(y_test_conf,axis=1)

        con_mat = tf.math.confusion_matrix(labels=y_test_conf, predictions=y_pred).numpy()
        #print(con_mat)
        acc = model2.evaluate(X_test, y_test)[1]
        cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = con_mat)
        cm_display.plot(cmap='Oranges')
        plt.show()
        print('accuracy : ',acc)
        max_acc = max(acc,max_acc)
        print("\n\n\n\n\n\n")
print("max acc with 3 layer networks is ", max_acc)
```
